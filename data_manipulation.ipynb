```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(
    "Untitled spreadsheet - Sheet1.csv"
)
```

```python
developers = data.groupby(['Original Developer']).size().sort_values(ascending=False)
active_developers = developers[developers > 10]
```

```python
data = data.rename(columns= {'Unnamed: 0':'Year'})
```

```python
active_developer_projects = data[data['Original Developer'].isin(active_developers.index)]
active_developer_projects
```

```python
active_developer_projects.to_csv('active_developer_projects.csv', index=False)
```
